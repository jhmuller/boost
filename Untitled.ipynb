```python
import yfinance as yf

def get_info_items(tick, items = None):
    res = ""
    items = ["longName", "marketCap", "bid", "ask","beta"]
    for item in items:
        try:
            res += f"{item}: {tick.info[item]}\n"
        except:
           res += f"{item}  NOT VALID\n"            
    return res

dflist = {}
symbols = ["HD", "LOW"] #['KO', 'PEP']
combo = pd.DataFrame()
for symbol in symbols:
    tick = yf.Ticker(symbol)
    
    # get stock info
    #res = get_info_items(tick)
    #print(res)
    #print()
    # get historical market data
    df = tick.history(period='5y')
    df = df[["Close"]]
    df.columns = [symbol]
    if combo.shape[0] == 0:
        combo = df
    else:
        combo = combo.join(df)
print(datetime.datetime.now())
```

```python
combo['diff'] = combo[symbols[0]] - combo[symbols[1]]
combo.plot(y="diff")
```
